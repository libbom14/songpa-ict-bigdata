```python
import seaborn as sns

# SSL 에러 해결
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

# 빈도표 작성 : value_counts 함수(pandas 데이터프레임)
iris = sns.load_dataset('iris')
iris.species.value_counts()
```

```python
# 빈도표 작성 : groupby함수, size함수(pandas 데이터프레임)
iris.groupby("species").size()
```

```python
# 데이터프레임의 양적(숫자형) 변수들에 대한 간단한 요약통계 량 값 출력
iris.describe()
```

```python
# include=‘object ’: 데이터프레임의 질적(문자형) 변수에 대한 요약통계량 값 출력
iris.describe(include='object')
```

```python
# include=‘all’ : 양적(숫자형)변수, 질적(문자형) 변수 모두에 대 한 요약통계량 값 출력
iris.describe(include='all')
```

```python
# 지정된 변수(양적, 질적)에 대한 요약통계량 값 출력
iris.sepal_length.describe()
```

```python
# 지정된 여러 개의 변수(양적, 질적)에 대하여 대한 요약통계 량 값 출력
iris.species.describe()
```

```python
iris[["sepal_length","petal_length"]].describe()
```

```python
# iris 품종(species)별로 sepal_length의 합계를 출력
iris.groupby("species")["sepal_length"].sum()
```

```python
# iris 품종(species)별로 sepal_width의 평균을 출력
iris.groupby("species")["sepal_width"].mean()
```

```python
# iris 품종(species)별로 petal_width의 표준편차를 출력
iris.groupby("species")["petal_width"].std()
```

```python
# species의 범주별로 sepal_length의 합계, sepal_width의 평균, petal_width의 최소값, 최대값을 출력
iris.groupby("species").agg({"sepal_length": "sum", "sepal_width": "mean", "petal_width": ["min", "max"]})
```

```python
# iris의 sepal_length 변수에 대하여 히스토그램 출력
# - column= : 히스토그램으로 출력할 변수 또는 변수 리스트 지정 
# - by = : 그룹변수 지정하여 그룹별 히스토그램을 그림
# - bins= : 구간의 개수를 지정, 기본값은 bins=10
# - grid=False : grid를 출력하지 않음
iris.hist(column="sepal_length")
iris.hist(column="sepal_length", bins=15)
iris.hist(column="sepal_length", by="species", grid=False)
```

```python
# ▶iris의 sepal_width 변수에 대하여 상자그림 출력
# - column= : 변수 이름이나 변수들의 리스트 저장
# - by= : 지정된 변수별로 상자그림을 그릴 경우 사용
# - grid= : 격자의 출력 여부 지정 기본값 : grid = True - vert=False : 수평상자그림 출력
iris.boxplot(column="sepal_width")
iris.boxplot(column="sepal_width", by="species", vert=False, grid=False)
```

```python
# stemgraphic 라이브러리로부터 stem_graphic 함수를 import한다
from stemgraphic import stem_graphic
# ▶ iris.sepal_length의 줄기-잎 그림 그리기
# - array, list, df 등
# - scale=줄기의 크기
stem_graphic(iris.sepal_length)
stem_graphic(iris.sepal_length, scale=1)
```
