```python
# Ïòà. (cars93.csv) cars93Ïùò ÏûêÎèôÏ∞® Ï§ëÎüâ(Weight)Í≥º ÏûêÎèôÏ∞® Ïó∞ÎπÑ(MPG.city) ÏÇ¨Ïù¥ÏóêÎäî ÏÑ†ÌòïÏ†Å Ïó∞Í¥ÄÏÑ±(ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ)Í∞Ä ÏûàÎäîÏßÄ Ïú†ÏùòÏàòÏ§Ä ùõº = 0.05ÏóêÏÑú ÏïåÏïÑÎ≥¥Ïûê.
# Í∞ÄÏÑ§
# - Í∑ÄÎ¨¥Í∞ÄÏÑ§ ùêª‚ÇÄ : ÏûêÎèôÏ∞® Ï§ëÎüâ(Weight)Í≥º ÏûêÎèôÏ∞® Ïó∞ÎπÑ(MPG.city) ÏÇ¨Ïù¥Ïóê ÏÑ†ÌòïÏ†Å Ïó∞Í¥ÄÏÑ±Ïù¥ Ï°¥Ïû¨ÌïòÏßÄ ÏïäÎäîÎã§
# - ÎåÄÎ¶ΩÍ∞ÄÏÑ§ ùêª‚ÇÅ : ÏûêÎèôÏ∞® Ï§ëÎüâ(Weight)Í≥º ÏûêÎèôÏ∞® Ïó∞ÎπÑ(MPG.city) ÏÇ¨Ïù¥Ïóê ÏÑ†ÌòïÏ†Å Ïó∞Í¥ÄÏÑ±Ïù¥ Ï°¥Ïû¨ÌïúÎã§
import pandas as pd
import numpy as np

cars93 = pd.read_csv("../day22/cars93.csv")
cars93.plot(kind='scatter', x='Weight', y='MPG.city')
np.corrcoef(cars93["Weight"], cars93["MPG.city"])
```

```python
# cars93Ïùò WeightÏôÄ MPG.city ÏÉÅÍ¥ÄÍ≥ÑÏàòÏôÄ ÏÇ∞Ï†êÎèÑÎ°ú Îëê Î≥ÄÏàòÍ∞ÑÏóê ÏñëÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÏûàÎã§Í≥† ÌåêÎã®ÎêòÏñ¥ ÏÉÅÍ¥ÄÎ∂ÑÏÑù Ïã§Ïãú
# pearsonr(scipy.stats Î™®Îìà) : Îëê ÏñëÏ†ÅÎ≥ÄÏàòÏóê ÎåÄÌïòÏó¨ ÏÉÅÍ¥ÄÎ∂ÑÏÑù Ïã§Ïãú
from scipy.stats import pearsonr

# ‚ñ∂ pearsonr(x= , y=) : Îëê Î≥ÄÏàòÍ∞ÑÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÏùò Ïú†Î¨¥Ïóê Í¥ÄÌïú Í≤ÄÏ†ï Ïã§Ïãú
# - x= , y= : Îëê Î≥ÄÏàòÎ•º ÏßÄÏ†ï
# - <Ïã§ÌñâÍ≤∞Í≥º> Ï£ºÏñ¥ÏßÑ Îëê Î≥ÄÏàò x, yÏóê ÎåÄÌïú ÌîºÏñ¥Ïä®Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàòÏôÄ Ïú†ÏùòÌôïÎ•†ÏùÑ Ï∂úÎ†•
pearsonr(cars93["Weight"], cars93["MPG.city"])
```

```python
# Ïòà. ÏãúÎ≥∏(seaborn)Ìå®ÌÇ§ÏßÄÏóê ÏûàÎäî tips Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóêÏÑú Î†àÏä§ÌÜ†ÎûëÏùò Ï¥ùÎß§Ï∂úÏï°(total_bill) Í≥º ÌåÅ(tip) ÏÇ¨Ïù¥Ïùò ÏÇ∞Ï†êÎèÑÎ•º Í∑∏Î¶¨Í≥† ÏÉÅÍ¥ÄÍ≥ÑÏàòÎ•º Î®ºÏ†Ä Íµ¨Ìï¥Î≥¥Ïûê. Îëê Î≥ÄÏàòÍ∞ÑÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÏûàÎäîÏßÄ Ïú†ÏùòÏàòÏ§Ä ùõº = 0.05ÏóêÏÑú ÌôïÏù∏ÌïòÏãúÏò§.
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

tips = sns.load_dataset("tips")

tips.plot(kind='scatter', x='tip', y='total_bill')
np.corrcoef(tips["tip"], tips["total_bill"])

pearsonr(tips["tip"], tips["total_bill"])
```
